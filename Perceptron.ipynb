# Perceptron Class

y=1, if $\sum_{i}w_i x_i >= b$   
y=0, otherwise

```python
from sklearn.metrics import accuracy_score
```

```python
class Perceptron:
    def __init__(self):
        self.w = None
        self.b = None
    
    def model(self,x):
        return 1 if (np.dot(self.w, x) >= self.b) else 0

    def predict(self, X):
        Y = []
        for x in X:
            result = self.model(x)
            Y.append(result)
        return np.array(Y)

    def fit(self, X, Y, epochs=10, learning_rate = 1):
        self.w = np.ones(X.shape[1])
        self.b = 0
        accuracy = []
        max_accuracy = 0
        wt_matrix = []
        test_accuracy=[]
        training_accuracy=[]
        for i in range(epochs):
            for x,y in zip(X,Y):
                Y_pred = self.model(x)
                if y == 1 and Y_pred == 0:
                    self.w = self.w + learning_rate*x
                    self.b = self.b + learning_rate*1
                elif y == 0 and Y_pred == 1:
                    self.w = self.w - learning_rate*x
                    self.b = self.b - learning_rate*1
            temp = accuracy_score(self.predict(X),Y)
            accuracy.append(temp)
            wt_matrix.append(self.w)
            if (temp > max_accuracy):
                max_accuracy = temp
                checkptw = self.w
                checkptb = self.b
            tempx,tempy=self.evolution()
            test_accuracy.append(tempy)
            training_accuracy.append(tempx)
        
        self.w = checkptw
        self.b = checkptb
        
        plt.plot(training_accuracy,color='blue',label='Training accuracy')
        plt.plot(test_accuracy, color='green',label='Test accuracy')
        plt.legend()
        plt.show()
        print("Best Result: ")
        Y_pred_train = perceptron.predict(X_train)
        print("Training accuracy:\t",accuracy_score(Y_pred_train,Y_train))
        Y_pred_test = perceptron.predict(X_test)
        print("Testing accuracy:\t",accuracy_score(Y_pred_test,Y_test))
        return np.array(wt_matrix)
    def evolution(self):
        Y_pred_train = perceptron.predict(X_train)
        train_ac=accuracy_score(Y_pred_train,Y_train)
        Y_pred_test = perceptron.predict(X_test)
        test_ac=accuracy_score(Y_pred_test,Y_test)
        return train_ac, test_ac
    

```

```python
perceptron = Perceptron()
```

```python
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
plt.plot(wt_matrix[-1,:])
plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_xlim(( 0, wt_matrix.shape[1]))
ax.set_ylim((-3000, 3000))

line, = ax.plot([], [], lw=2)
```

```python
def animate(i):
    x = list(range(wt_matrix.shape[1]))
    y = wt_matrix[i,:]
    line.set_data(x, y)
    return (line,)
```

```python
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=100, interval=100, blit=True)
```

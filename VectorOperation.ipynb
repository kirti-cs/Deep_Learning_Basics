Importing basic Libary

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Defining Function to plot VECTORS

```python
def plotVector(vect):
    color = ['r','g','b','y']
    i=0
    for x in vect:
        plt.quiver(x[0],x[1],x[2],x[3], scale_units='xy',angles='xy',scale=1,color=color[i%4])
        plt.xlim(-10,10)
        plt.ylim(-10,10) 
        i+=1
    plt.show()
```

Intializing a List of vector with Numpy asarray, because numpy support +, - , * , / basic operation on the go

```python
A = [np.asarray([0,0,3,4]), np.asarray([0,0,-3,4]),np.asarray([0,0,4,2]),np.asarray([0,0,-3,-4])]
```

# ADD Operation:
You can change the value of this vector from A List

```python
plotVector([A[0],A[1],A[0]+A[1]])
```

# Substruct Operation:
You can change the value of this vector from A List

```python
plotVector([A[0],A[1],A[0]-A[1]])
```

# Multiplication Operation:
You can change the value of this vector from A List

```python
plotVector([A[0],A[1],A[0]*A[2]])
```

# Vector Dot Product
You can change the value of this vector from A List

```python
a = np.asarray([3,4])
b = np.asarray([4,-1])
```

Dot Product Basic Formula

$\vec{a}\cdot\vec{b}=|\vec{a}| |\vec{b}| \cos(\theta) =a_x b_x + a_y b_y$

```python
a_dot_b = np.dot(a,b)
print(a_dot_b)
```

$a_b = |\vec{a}| \cos(\theta) = |\vec{a}|\frac{\vec{a}\cdot\vec{b}}{|\vec{a}||\vec{b}|}=\frac{\vec{a}\cdot\vec{b}}{|\vec{b}|} $

```python
a_b = np.dot(a,b)/np.linalg.norm(b)
print(a_b)
```

$\vec{a_b}=a_b\hat{b}=a_b \frac{\vec{b}}{|\vec{b}|}$

```python
vec_a_b = a_b/np.linalg.norm(b)*b
print(vec_a_b)
```

Now For Graphically Explanation we are projecting Red Vector on Blue Vector
